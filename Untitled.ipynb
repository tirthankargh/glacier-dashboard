```python
import geopandas as gpd

# Load shapefile
glaciers = gpd.read_file("/Users/tirthankar/Downloads/RGI2000-v7.0-G-14_south_asia_west/RGI2000-v7.0-G-14_south_asia_west.shp")

print(type(glaciers))
print(glaciers.head())

# Reproject to WGS84 (Leaflet requirement)
glaciers = glaciers.to_crs(epsg=4326)

# Keep only useful columns (edit this!)
keep_cols = ["rgi_id", "area_km2", "zmean_m"]
glaciers = glaciers[keep_cols]

# Rename for cleaner frontend use
glaciers = glaciers.rename(columns={
    "rgi_id": "id",
    "area_km2": "area"
})

# Simplify geometry (VERY important for performance)
glaciers["geometry"] = glaciers["geometry"].simplify(
    tolerance=0.001,  # adjust if needed
    preserve_topology=True
)

# Save as GeoJSON
glaciers.to_file("data/glaciers.geojson", driver="GeoJSON")

print("✅ Glacier GeoJSON created")
```

```python
import geopandas as gpd

# Load
gdf = gpd.read_file("/Users/tirthankar/Downloads/RGI2000-v7.0-G-14_south_asia_west/RGI2000-v7.0-G-14_south_asia_west.shp")

# Ensure CRS is WGS84
gdf = gdf.to_crs(epsg=4326)

# --- DROP Z DIMENSION (IMPORTANT) ---
gdf["geometry"] = gdf["geometry"].apply(
    lambda geom: geom if geom is None else geom.__class__([(x, y) for x, y, *_ in geom.exterior.coords])
    if geom.geom_type == "Polygon"
    else geom
)

# --- Keep only essential columns ---
gdf = gdf[["rgi_id", "area_km2", "cenlat", "cenlon", "geometry"]]

# Rename for frontend
gdf = gdf.rename(columns={
    "rgi_id": "id",
    "area_km2": "area"
})

# --- Simplify geometry ---
gdf["geometry"] = gdf["geometry"].simplify(
    tolerance=0.001,
    preserve_topology=True
)

# --- Save ---
gdf.to_file("data/glaciers.geojson", driver="GeoJSON")

print("✅ Glacier GeoJSON ready")
```

```python
import geopandas as gpd

basins = gpd.read_file("/Users/tirthankar/PhD/OGGM Modelling/Data/L8_basins.shp")

# Reproject
basins = basins.to_crs(epsg=4326)

# Keep minimal attributes
keep_cols = ["HYBAS_ID", "geometry"]  # adjust based on your file
basins = basins[keep_cols]

basins = basins.rename(columns={
    "HYBAS_ID": "id"
})

# Simplify (slightly less aggressive than glaciers)
basins["geometry"] = basins["geometry"].simplify(
    tolerance=0.005,
    preserve_topology=True
)

# Save
basins.to_file("data/basins.geojson", driver="GeoJSON")

print("✅ Basin GeoJSON created")
```
